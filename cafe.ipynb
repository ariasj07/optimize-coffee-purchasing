### Café óptimo para comprar 10000 gramos (10kg)
Café que salga más barato para realizar una compra de 10000 gramos de. No puede ser menor a 10000 gramos (10kg)

```python
import pandas as pd
df = pd.read_json("cafe.json") # Load the data scraped from https://www.walmart.co.cr/
import re
```

```python
pattern = r'\b\d+\s?[gG][rR]?\b'


df["product_name"].str.lower()
df["price_clean"] = df["price"].apply(lambda price: int(price.replace(".", "").replace("₡", "")))
df["grams"] = df["product_name"].apply(lambda name: re.search(pattern, name).group(0))
df["grams"] = df["grams"].apply(lambda q: int(re.sub(r"[a-zA-Z]", "", q.strip())))
counter = 1
grams = df["grams"]
tries = []
counter = 1

for g in grams:
    while True:
        if g * counter >= 10000:
            tries.append(counter)
            counter = 1
            break
        else:
            counter = counter + 1

def nombre():
    print("i")
# df = df.drop("link", axis=1).drop("image_src", axis=1).drop("product_promo", axis=1) <--- Deleted unnecessary columns, run it when loading the data
df["purchases"] = tries
df["total_grams"] = df["purchases"] * df["grams"]
df["bought_amount"] = df["purchases"] * df["price_clean"]
df.sort_values(by="bought_amount", ascending=True).head(10)
```
